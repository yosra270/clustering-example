## Clustering

### Hierarchical clustering

#### Agglomerative hierarchical clustering

```python
# 1st method : we cluster the dataset by choosing the minimum distance between two clusters

import pandas
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

Z = linkage(fromage,method='ward', metric='euclidean') # Apply the clustering algorithm

# Display the dendrogram
plt.figure(figsize=(15,10)) # Increase dendrogram's resolution
plt.title('Cheese dataset Dendrogram')
dendrogram(Z,labels=fromage.index,orientation='left',color_threshold=255)
plt.show()

# Clustering the dataset
clusters = fcluster(Z, t = 255,criterion='distance') # Divide the dataset into clusters whose distance does not exceed 255
print("For a minimum distance of 255 between 2 clusters, we obtain ", len(set(clusters)), " clusters.\n\n")

# Display the cluster of each sample
idg = np.argsort(clusters)
print(pandas.DataFrame(fromage.index[idg],clusters[idg]))
```

```python
# 2nd method : we cluster the dataset by choosing the number of clusters

from sklearn.cluster import AgglomerativeClustering # AgglomerativeClustering uses linkage distance to recursively merges pair of clusters

np.random.seed(0)
agg_c = AgglomerativeClustering(n_clusters=4) # We need to provide the number of clusters beforehand
agg_c.fit(fromage)
idk_ = np.argsort(agg_c.labels_)

print(pd.DataFrame(fromage.index[idk_],agg_c.labels_[idk_]))
```

**Problem :** We need the determine the best number of clusters.

**Solution :** We will need to use the 'silhouette' metric. (We used this metric to determine the best k in k-Means).

#### Divisive hierarchical clustering

```python
import pandas as pd
import numpy as np
from sklearn import cluster
from sklearn import metrics

def divisive_clustering_using_kmeans(oldDataset, step_nbre = 0):
    if len(oldDataset)==1:
        step_nbre = step_nbre+1
        print("\n")
        print("######  step final ",step_nbre," :  ######")
        print(" \n".join(pd.DataFrame(oldDataset).index))
        return
    else:
        # Prepare the k-Means algorithm with k = 2 to separate the dataset into two new clusters
        km = cluster.KMeans(n_clusters = 2)
        km.fit(oldDataset)
    
        # Separate dataset into two clusters
        tab0=[] # Rows numbers of the first cluster
        tab1=[] # Rows numbers of the second cluster
        for i  in np.arange(len(km.labels_)):
            if km.labels_[i]==0:
                tab0.append(i)
            elif km.labels_[i]==1:
                tab1.append(i)

        # Extract the samples of each cluster
        sub0=oldDataset.iloc[tab0]
        sub1=oldDataset.iloc[tab1]

        step_nbre=step_nbre+1

        # Display the two clusters
        print("\n")
        print("######  step  ",step_nbre," :  ######")
        print("\n")
        print("first cluster for this step  : \n")
        print(" \n".join(pd.DataFrame(sub0).index))
        print("\n")
        print("second cluster for this step : \n")
        print(" \n".join(pd.DataFrame(sub1).index))
        
        # Keep splitting the resulting clusters
        divisive_clustering_using_kmeans(sub0,step_nbre)
        divisive_clustering_using_kmeans(sub1,step_nbre)
```

```python
divisive_clustering_using_kmeans(fromage)
```

### Centroid-based clustering

#### k-Means clustering

```python
import numpy as np
from sklearn import cluster

# np.random.seed(0) makes the random numbers predictable
np.random.seed(0)

# Initialize k-Means algorithm with k = 4 (number of clusters to form as well as the number of centroids to generate)
kmeans = cluster.KMeans(n_clusters = 4)
kmeans.fit(fromage)

# Sort cheese labels by index
idk = np.argsort(kmeans.labels_)

print(pd.DataFrame(fromage.index[idk],kmeans.labels_[idk]))
```

```python
# Transform 'fromge' to a cluster-distance space

kmeans.transform(fromage)
```

```python
# Cluster centers
print("Cluster centers :\n\n",kmeans.cluster_centers_)
```

**Problem :** k is chosen by the user (in reality it should be chosen by an expert).

**Solution :** to determine the best k value for a good clustering we can use the 'silhouette' metric.

```python
# Determine best value of k 

from sklearn import metrics

res = np.arange(8, dtype ="double") 
for k in np.arange(8):
    km = cluster.KMeans(n_clusters = k+2)
    km.fit(fromage)
    res[k] = metrics.silhouette_score(fromage,km.labels_)

# Plotting the result
import matplotlib.pyplot as plt
plt.title("silhouette")
plt.xlabel("number of clusters")
plt.ylabel("performance")
plt.plot(np.arange(2,10,1),(res))
plt.show
```

The highest score is around 0.5 for a number of clusters of 2. However, one should avoid using a value of k of 2 (this is a borderline case). We therefore choose the number 5 which has the best score according to this metric.

If we choose a k that is:

    * very low => risk that elements belonging to the same cluster present dissimilarities.
    * very high => risk that elements not belonging to the same cluster are similar.


The choice of the number of clusters will depend on the confidence of the user and the case study:

    * less confident user => more specific clusters
    * more confident user => more generic clusters


### Correspondence between the clusters generated by the different algorithms

```python
# Correspondence between clusters generated by k-Means and AgglomerativeClustering
pd.crosstab(kmeans.labels_,agg_c.labels_, rownames=['k-Means'], colnames=['AgglomerativeClustering',])
```

* Cluster 0 of k-Means <=> Cluster 2 of AgglomerativeClustering
* Cluster 1 of k-Means <=> Cluster 0 of AgglomerativeClustering
* Cluster 2 of k-Means <=> Cluster 3 of AgglomerativeClustering
* Cluster 3 of k-Means <=> Cluster 1 of AgglomerativeClustering

With 2 exceptions : 4 samples of k-means 4th cluster  are not part of AgglomerativeClustering 2nd cluster and an elemnt of k-means 1st cluster  is not part of AgglomerativeClustering 3rd cluster.

```python
# Correspondence between clusters generated by k-Means and linkage
pd.crosstab(kmeans.labels_,clusters, rownames=['k-Means'], colnames=['Ward linkage',])
```

* Cluster 0 of k-Means <=> Cluster 4 of Ward linkage
* Cluster 1 of k-Means <=> Cluster 2 of Ward linkage
* Cluster 2 of k-Means <=> Cluster 1 of Ward linkage
* Cluster 3 of k-Means <=> Cluster 3 of Ward linkage

With 2 exceptions : 4 samples of k-means 4th cluster are not part of Ward linkage 3rd cluster and an element of k-means 1st cluster is not part of Ward linkage 4th cluster.


```python
# Correspondence between clusters generated by linkage and AgglomerativeClustering
pd.crosstab(agg_c.labels_,clusters, rownames=['Ward linkage'], colnames=['AgglomerativeClustering',])
```

* Cluster 0 of Ward linkage <=> Cluster 2 of AgglomerativeClustering
* Cluster 1 of Ward linkage <=> Cluster 3 of AgglomerativeClustering
* Cluster 2 of Ward linkage <=> Cluster 4 of AgglomerativeClustering
* Cluster 3 of Ward linkage <=> Cluster 1 of AgglomerativeClustering

There is a strong correspondence between the clusters generated by both algorithms.
